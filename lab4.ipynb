### Выполнение:

```python
import random
import time


def generate_array(count):
   array = [int(random.randint(-999, 1000)) for i in range(count)]
   return array

```

***
### Бинарный поиск

```python
generated_array = generate_array(20)
generated_array.sort()
print(generated_array)

def binary_search(arr, target):
    left = 0
    right = len(arr) - 1

    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid  
        elif arr[mid] < target:
            left = mid + 1  
        else:
            right = mid - 1  

    return -1  
target = int(input())
start_time = time.time()
result = binary_search(generated_array, target)
print("--- {0} ms ---".format(time.time() - start_time), "Бинарный поиск")
if result != -1:
    print("Элемент", target, "найден по индексу", result)
else:
    print("Элемент", target, "не найден в массиве")
```

### Бинарное дерево

```python
class Node:
    def __init__(self, key, value):
        self.left = None
        self.right = None
        self.key = key
        self.value = value

def insert(root, key, value):
    if root is None:
        return Node(key, value)
    if key < root.key:
        root.left = insert(root.left, key, value)
    elif key > root.key:
        root.right = insert(root.right, key, value)
    else:
        root.value = value
    return root

def min_value_node(node):
    current = node
    while current.left is not None:
        current = current.left
    return current

def output(root):
    if root:
        output(root.left)
        print(f"Key: {root.key}, Value: {root.value}")
        output(root.right)

def search(root, key):
    if root is None or root.key == key:
        return root.value
    if root.key < key:
        return search(root.right, key)
    return search(root.left, key)

def delete_node(root, key):
    if root is None:
        return root
    if key < root.key:
        root.left = delete_node(root.left, key)
    elif key > root.key:
        root.right = delete_node(root.right, key)
    else:
        if root.left is None:
            return root.right
        elif root.right is None:
            return root.left
        temp = min_value_node(root.right)
        root.key = temp.key
        root.right = delete_node(root.right, temp.key)
    return root

def array_to_binary_tree(arr, start, end):
    if start > end:
        return None

    mid = (start + end) // 2
    root = Node(arr[mid], mid)

    root.left = array_to_binary_tree(arr, start, mid - 1)
    root.right = array_to_binary_tree(arr, mid + 1, end)

    return root

array = generate_array(20)
array.sort()
print(array)

target = int(input())
start_time = time.time()
root = array_to_binary_tree(array, 0, len(array) - 1)
result = search(root, target)
print("--- {0} ms ---".format(time.time() - start_time), "Бинарное дерево поиска")
print("Элемент", target, "найден по индексу", result)
```
